#### Examine Date of Birth

```python
std_file['date_of_birth'] = std_file['date_of_birth'].str.strip()
```

```python
# possible_formats = ['%d-%m-%y', '%d-%m-%']
from datetime import datetime
import re
possible_formats = []
Years = ['%y', '%Y']
Months = ['%m', '%b', '%B']
Day = ['%d']
for year in Years:
    for month in Months:
        for day in Day:
            possible_formats.append(f"{day} {month} {year}")
```

```python
possible_formats
```

```python
def getDateFromString(date_str):
    date_str = re.sub(r'(-|/)', ' ', date_str)
    for date_format in possible_formats:
        try:
            return datetime.strptime(date_str, date_format).date()
        except:
            pass
    else:
        print("Unable to parse date", date_str)
```

```python
std_file['date_of_birth'] = std_file['date_of_birth'].apply(lambda x:getDateFromString(x))
```
